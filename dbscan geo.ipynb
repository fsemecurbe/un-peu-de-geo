```python
import numpy as np
import pandas as pd
import geopandas as gpd

from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

## Application de la dbscan aux données géographiques

```python
marne = pd.read_csv("marne.csv")
manche = pd.read_csv("manche.csv")
lauragais = pd.read_csv("lauragais.csv")
```

Je calcule la dbscan avec eps = 2OO m et 10 voisins pour la marne

```python
db = DBSCAN(eps=200, min_samples=10).fit(marne)
```

Je rajoute le résultat dans la base marne

```python
marne['monstre'] = (db.labels_ > 0) * 1
```

Pour créer un fichier géographique

```python
smarne = gpd.GeoDataFrame(marne, geometry=gpd.points_from_xy(marne.x, marne.y),crs="EPSG:2154")
```

Pour afficher ma carte

```python
smarne.explore('monstre')
```
